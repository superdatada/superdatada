# 1. Prep 

**Instalacin e importacin de libreras**

```python
# Bibliotecas estndar
import re
import time

# Bibliotecas externas
import billboard as bd
import lyricsgenius
import numpy as np
import pandas as pd
import torch
from bertopic import BERTopic
from bertopic.representation import KeyBERTInspired
from fuzzywuzzy import fuzz
from langdetect import detect
from sklearn.decomposition import LatentDirichletAllocation, NMF
from transformers import BertModel, BertTokenizer
```

**Autenticacin de la API**


Para poder acceder a las letras de canciones desde Genius, es necesario autenticarse usando una clave API **(api_key)**. Esta clave es un cdigo nico que obtienes al registrarte en la [plataforma de Genius](https://genius.com/developers).

```python
# Define la autenticacin para la API de Genius con la clave proporcionada
# Configura opciones para saltar canciones no oficiales, excluir remixes y versiones en vivo,
# y eliminar los encabezados de seccin en las letras. Tambin establece un tiempo mximo de espera.
api_key = ""
genius = lyricsgenius.Genius(
    api_key,
    skip_non_songs=False,
    excluded_terms=["(Remix)", "(Live)"],
    remove_section_headers=True,
    timeout=10,
)
```

**Funciones auxiliares**


En esta seccin encontrars funciones que nos ayudan a descargar canciones y letras, limpiar y normalizar los textos, detectar idiomas y comparar nombres de artistas con tcnicas de coincidencia difusa. Estas herramientas son clave para preparar los datos y facilitar el anlisis musical.

```python
def obtener_canciones_por_ao(ao_inicial: int) -> pd.DataFrame:
    """
    Obtiene canciones del Billboard Hot 100 para dos aos consecutivos a partir del ao dado.

    Args:
        ao_inicial (int): Ao desde el cual comenzar a obtener datos.

    Returns:
        pd.DataFrame: DataFrame con columnas ['artists', 'weeks', 'songs', 'years'].
    """
    canciones = []
    artistas = []
    semanas = []
    aos = []

    meses = np.char.zfill(np.arange(1, 13).astype(str), 2)  # '01', '02', ..., '12'
    aos_consulta = np.arange(ao_inicial, ao_inicial + 2)

    for ao in aos_consulta:
        for mes in meses:
            fecha = f"{ao}-{mes}-15"
            chart = bd.ChartData("hot-100", date=fecha)
            for entrada in chart:
                canciones.append(entrada.title)
                artistas.append(entrada.artist)
                semanas.append(entrada.weeks)
                aos.append(ao)

    df = pd.DataFrame(
        {"artists": artistas, "weeks": semanas, "songs": canciones, "years": aos}
    )
    return df


def obtener_letras(df: pd.DataFrame) -> pd.DataFrame:
    """
    Busca y descarga las letras de canciones usando la API de Genius.

    Args:
        df (pd.DataFrame): DataFrame con columnas 'songs' y 'artists_genius'.

    Returns:
        pd.DataFrame: DataFrame con letras y datos asociados.
    """
    letras = []
    artistas_encontrados = []
    titulos_encontrados = []
    decadas = []
    busqueda_cancion = []
    busqueda_artista = []

    for idx in range(df.shape[0]):
        try:
            print(
                f"Buscando letra {idx+1}/{df.shape[0]}: {df.iloc[idx]['songs']} - {df.iloc[idx]['artists_genius']}"
            )
            resultado = genius.search_song(
                df.iloc[idx]["songs"], df.iloc[idx]["artists_genius"]
            )
            letras.append(resultado.lyrics if resultado else "not found")
            artistas_encontrados.append(
                resultado.artist if resultado else df.iloc[idx]["artists_genius"]
            )
            titulos_encontrados.append(
                resultado.title if resultado else df.iloc[idx]["songs"]
            )
        except Exception:
            letras.append("not found")
            artistas_encontrados.append(df.iloc[idx]["artists_genius"])
            titulos_encontrados.append(df.iloc[idx]["songs"])

        decadas.append(df.iloc[idx]["decade"] if "decade" in df.columns else None)
        busqueda_cancion.append(df.iloc[idx]["songs"])
        busqueda_artista.append(df.iloc[idx]["artists_genius"])
        time.sleep(0.3)

    resultado_df = pd.DataFrame(
        {
            "lyrics": letras,
            "artists_genius_found": artistas_encontrados,
            "song_genius_found": titulos_encontrados,
            "decade": decadas,
            "song_search": busqueda_cancion,
            "artist_search": busqueda_artista,
        }
    )
    return resultado_df


def transformar_nombre_artista(nombre: str) -> str:
    """
    Normaliza el nombre del artista para facilitar comparaciones.

    Pasos:
    - Pasa a minsculas.
    - Cambia 'x' por '&'.
    - Elimina colaboraciones ('featuring', 'feat.', etc.).
    - Elimina contenido entre parntesis y corchetes.

    Args:
        nombre (str): Nombre original del artista.

    Returns:
        str: Nombre normalizado.
    """
    nombre = nombre.lower()
    nombre = re.sub(r"\bx\b", "&", nombre)

    for palabra in ["featuring", "with", "feat.", "duet", "(", "["]:
        nombre = re.split(re.escape(palabra), nombre, flags=re.IGNORECASE)[0].strip()

    nombre = re.sub(r"\([^()]*\)", "", nombre)
    nombre = re.sub(r"\[[^\[\]]*\]", "", nombre)

    return nombre.strip()


def limpiar_letras(letra: str) -> str:
    """
    Limpia la letra eliminando texto innecesario y caracteres no deseados.

    Args:
        letra (str): Letra original.

    Returns:
        str: Letra limpia y normalizada.
    """
    letra_limpia = re.sub(r".*Lyrics", "", letra, flags=re.DOTALL)
    letra_limpia = re.sub(r"[0-9\W_]+", " ", letra_limpia)
    letra_limpia = re.sub(
        r"chorus|instrumental|you might also like|embed",
        "",
        letra_limpia,
        flags=re.IGNORECASE,
    )
    letra_limpia = " ".join(letra_limpia.split())
    return letra_limpia


def detectar_idioma(texto: str) -> str:
    """
    Detecta el idioma del texto dado.

    Args:
        texto (str): Texto a analizar.

    Returns:
        str: Cdigo del idioma detectado o 'Unknown' si falla.
    """
    try:
        return detect(texto)
    except Exception:
        return "Unknown"


def comparar_artistas_fuzzy(fila: dict) -> int:
    """
    Calcula la similitud entre dos nombres de artistas usando comparacin difusa.

    Args:
        fila (dict): Diccionario con las claves 'artists_genius_found' y 'artist_search'.

    Returns:
        int: Puntuacin de similitud entre 0 y 100.
    """
    try:
        return fuzz.token_set_ratio(
            fila["artists_genius_found"].lower(), fila["artist_search"].lower()
        )
    except Exception:
        return 0
```

# 2. Data 

### Canciones
 Ten paciencia! Estamos recopilando 100 canciones cada dos semanas desde 1960. Si un mes tiene 4 semanas, eso son 48 semanas al ao. Cada quincena equivale a 24 periodos por ao. Y desde 1960 hasta 2023? Haz los clculos!

```python
# Lista de aos iniciales para obtener canciones
aos_inicio = [1960, 1970, 1980, 1990, 2000, 2010, 2022, 2023]

# Obtener y concatenar los DataFrames de canciones para cada ao
df_canciones = pd.concat([obtener_canciones_por_ao(ao) for ao in aos_inicio])

# Eliminar duplicados considerando artista y cancin
df_canciones = df_canciones.drop_duplicates(subset=["artists", "songs"])

# Crear columna 'decade' redondeando hacia abajo a la dcada ms cercana
df_canciones["decade"] = (df_canciones["years"] // 10) * 10

# Volver a eliminar duplicados y agrupar para obtener mximo de semanas por grupo
df_canciones = (
    df_canciones.drop_duplicates()
    .groupby(["artists", "songs", "decade"], as_index=False)["weeks"]
    .max()
)

# Filtrar el DataFrame para tener un nmero similar de canciones por dcada
df_filtrado = pd.concat(
    [
        df_canciones[(df_canciones["decade"] < 1990) & (df_canciones["weeks"] >= 3)],
        df_canciones[
            (df_canciones["decade"] >= 1990)
            & (df_canciones["decade"] < 2020)
            & (df_canciones["weeks"] > 2)
        ],
        df_canciones[df_canciones["decade"] == 2020],
    ],
    ignore_index=True,
)
```

### Lyrics

#### Pasos del procesamiento 

1. **Limpieza y normalizacin de nombres de artistas**   
   Para asegurarnos de que los nombres de los artistas coincidan correctamente con los registros de la API de Genius, realizamos una limpieza cuidadosa. Esto incluye eliminar partes irrelevantes como colaboraciones, smbolos extraos o contenidos entre parntesis. As, evitamos confusiones y mejoramos la calidad de la bsqueda.

2. **Obtencin de letras de canciones**   
   A partir de los nombres limpios de artistas y ttulos de canciones, consultamos la API de Genius para descargar las letras. En total, procesamos ms de 10,700 canciones, un trabajo que puede tardar entre 3 y 4 horas, pero que vale completamente la espera!

3. **Preprocesamiento y filtrado de letras**   
   Limpiamos las letras para eliminar textos innecesarios como instrucciones, etiquetas, nmeros o caracteres especiales. Adems, aplicamos una comparacin inteligente para validar que el artista encontrado sea el correcto y detectamos automticamente el idioma de cada cancin para un anlisis ms preciso.

   Debido a que la API de Genius no es perfecta, implementamos estas etapas adicionales de procesamiento del lenguaje natural (NLP) para mejorar la calidad y precisin de los datos:

   - **Identificacin de idioma**:  
     Utilizamos [`langdetect`](https://pypi.org/project/langdetect/), una librera sencilla y efectiva para detectar el idioma del texto.

   - **Tokenizacin, eliminacin de stopwords y lematizacin/raz**:  
     Para esto recurrimos a libreras robustas como:
     - [`NLTK`](https://www.nltk.org/): un toolkit completo para tareas de procesamiento de texto.  
     - [`spaCy`](https://spacy.io/): una biblioteca de NLP industrial que ofrece tokenizacin, lematizacin, etiquetado gramatical y ms.

---

Comencemos esta aventura musical! 



**(1) Limpieza y normalizacin de nombres de artistas**  

```python
# Crear una nueva columna 'artists_genius' con los nombres de artistas normalizados
df_filtrado["artists_genius"] = df_filtrado["artists"].map(transformar_nombre_artista)
df_filtrado
```

**(2) Obtencin de letras de canciones**  

```python
# Obtener las letras de las canciones a partir del DataFrame de canciones
df_lyr = obtener_letras(df_filtrado)
df_lyr
```

**(3) Preprocesamiento y filtrado de letras**   

```python
# Filtrar letras que fueron encontradas y no estn vacas
df = df_lyr[(df_lyr["lyrics"] != "not found") & (df_lyr["lyrics"] != "")].copy()

# Limpiar las letras de canciones usando la funcin definida
df["clean_lyrics"] = df["lyrics"].map(limpiar_letras)

# Detectar el idioma de las letras limpias
df["language"] = df["clean_lyrics"].map(detectar_idioma)

# Calcular el puntaje de similitud entre el artista encontrado y el buscado
df["similarity_score"] = df.apply(comparar_artistas_fuzzy, axis=1)

# Filtrar resultados con alta similitud y idioma vlido (excluyendo estonio y desconocido)
df = df[
    (df["similarity_score"] > 64)
    & (df["language"] != "et")
    & (df["language"] != "Unknown")
].copy()
```

# 3. Modelo 

Usamos **mBERT** (Multilingual BERT), una versin multilinge del modelo BERT, que es altamente eficiente para tareas de procesamiento de lenguaje natural (NLP). mBERT ha sido entrenado con datos en varios idiomas, lo que lo hace ideal para trabajar con textos multilinges.

En este proyecto, lo utilizamos para **modelado de temas** (topic modeling) en un corpus multilinge, ayudndonos a identificar patrones y temas recurrentes en las letras de canciones a lo largo del tiempo.


## 1960

```python
%%time  # Mide el tiempo de ejecucin de esta celda

# Crear el modelo de representacin basado en KeyBERT
representation_model = KeyBERTInspired()  # KeyBERTInspired es un modelo de representacin basado en palabras clave

# Crear el modelo de BERTopic con parmetros ajustados
model = BERTopic(
    verbose=True,              # Muestra informacin detallada durante el entrenamiento
    min_topic_size=10,         # Nmero mnimo de documentos por tema
    language='multilingual',   # Especifica que el modelo debe ser capaz de manejar mltiples idiomas
    representation_model=representation_model  # Usamos el modelo de representacin personalizado
)

# Entrenamos el modelo de BERTopic usando las letras de canciones de la dcada de 1960
headline_topics, _ = model.fit_transform(df[df.decade == 1960].clean_lyrics.values)
```

```python
# Obtener la informacin de los temas generados por el modelo
freq = model.get_topic_info()

# Imprimir el nmero total de temas identificados
print("Nmero de temas encontrados: {}".format(len(freq)))

freq.head()
```

```python
# Visualizar los 10 principales temas en un grfico de barras
# top_n_topics=10 especifica que queremos ver los 10 temas ms frecuentes
model.visualize_barchart(top_n_topics=10)
```

```python
# Visualizar la jerarqua de los temas en un grfico de rbol
model.visualize_hierarchy(top_n_topics=30)
```
